```python
import pandas as pd
```

```python
df=pd.read_excel("Bank Data Analystics.xlsx")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Dateof Birth']=pd.to_datetime(df['Dateof Birth'],errors='coerce')
```

```python
df['Closed Date']=pd.to_datetime(df['Closed Date'],errors='coerce')
```

```python
df['Disbursement Date']=pd.to_datetime(df['Disbursement Date'],errors='coerce')
```

```python
df['NextMeetingDate']=pd.to_datetime(df['NextMeetingDate'],errors='coerce')
```

```python
max_date=df['Closed Date'].max()
df['Closed Date']=df['Closed Date'].fillna(max_date)
df['Closed Date']=df['Closed Date'].dt.date
```

```python
max_date=df['Disbursement Date'].max()
df['Disbursement Date']=df['Disbursement Date'].fillna(max_date)
df['Disbursement Date']=df['Disbursement Date'].dt.date
```

```python
max_date=df['NextMeetingDate'].max()
df['NextMeetingDate']=df['NextMeetingDate'].fillna(max_date)
df['NextMeetingDate']=df['NextMeetingDate'].dt.date
```

```python
max_date=df['Dateof Birth'].max()
df['Dateof Birth']=df['Dateof Birth'].fillna(max_date)
df['Dateof Birth']=df['Dateof Birth'].dt.date
```

```python
df.rename(columns={'Grrade':'Grade'},inplace=True)
```

```python
df.rename(columns={'Age _T':'Age'},inplace=True)
```

```python
max_grade=df['Grade'].mode()[0]
df['Grade']=df['Grade'].fillna(max_grade)
```

```python
mean_age=df['Age'].mean()
df['Age']=df['Age'].fillna(mean_age)
```
